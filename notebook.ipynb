```python
X = df[["Average Cost for two", "Has Table booking", "Has Online delivery", "Price range"]]
```

```python
X
```

```python
y = df[["Aggregate rating"]]
```

```python
y
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
X = X.copy()
X["Has Table booking"] = labelencoder.fit_transform(X["Has Table booking"])
```

```python
X
```

```python
labelencoder.classes_
```

```python
X["Has Online delivery"] = labelencoder.fit_transform(X["Has Online delivery"])
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
def modelresults(predictions):
    print("Mean absolute error on model is: {}".format(mean_absolute_error(y_test, predictions)))
    print("Root mean squared error on model is: {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictionfromlr=lr.predict(X_test)
```

```python
modelresults(predictionfromlr)
```

```python
predictionfromlr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C" : [0.1,0.5], "kernel" : ["linear", "poly"], "degree" : [2]}
```

```python
param_gridsvr
```

```python
gridsvr = GridSearchCV(svrmodel,param_gridsvr)
gridsvr.fit(X_train,y_train)
```

```python
predgridsvr=gridsvr.predict(X_test)
```

```python
modelresults(predgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    'max_depth': [10, 20],
    'min_samples_split': [2,5],
    'min_samples_leaf': [1, 2]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=treemodel, param_grid=param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
treepredictions=grid_tree.predict(X_test)
```

```python
modelresults(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
gridrfr = GridSearchCV(rfrmodel, param_grid=param_grid)
```

```python
gridrfr.fit(X_train,y_train)
```

```python
randompredictions = gridrfr.predict(X_test)
```

```python
modelresults(randompredictions)
```

```python
gridrfr.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {'n_neighbors': [9, 11, 14, 15,16]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```

```python
knn_grid_search.fit(X_train, y_train)
```

```python
knnpred = knn_grid_search.predict(X_test)
```

```python
modelresults(knnpred)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada_param_grid = {'n_estimators': [50, 100, 200],
              'learning_rate': [ 0.1,0.5, 1]}
ada_grid = GridSearchCV(ada, ada_param_grid)  
ada_grid.fit(X_train, y_train)
```

```python
adsapred = ada_grid.predict(X_test)
```

```python
modelresults(adsapred)
```
